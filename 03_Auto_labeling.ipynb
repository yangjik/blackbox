```python
import cv2

# 이미지 로드
image = cv2.imread('./img/7ACb7kLtZtI_11.jpg')

# Contour 찾기
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
contours, _ = cv2.findContours(threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 두 개체 중심 좌표 계산
object1 = contours[0]
M1 = cv2.moments(object1)
object1_center = (int(M1['m10'] / M1['m00']), int(M1['m01'] / M1['m00']))

object2 = contours[1]
M2 = cv2.moments(object2)
object2_center = (int(M2['m10'] / M2['m00']), int(M2['m01'] / M2['m00']))

# 거리 계산
distance = ((object1_center[0] - object2_center[0])**2 + (object1_center[1] - object2_center[1])**2)**0.5

# 거리 출력
print(f"두 개체 사이의 거리: {distance} 픽셀")
```
