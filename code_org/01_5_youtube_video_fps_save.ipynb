### 테스트 코드

```python
video_path = './video/12.17/INF/20210810_011131_INF_2.avi'
file = video_path.split('/')[-1].split('.')[0]

video = cv2.VideoCapture(video_path)

if not video.isOpened():
    print('해당 파일 열수 없습니다. ', video_path)
    sys.exit()

# 동영상 정보 확인
video_width = video.get(cv2.CAP_PROP_FRAME_WIDTH)       # 동영상 width
video_hight = video.get(cv2.CAP_PROP_FRAME_HEIGHT)      # 동영상 hight
video_count = video.get(cv2.CAP_PROP_FRAME_COUNT)       # 총 프레임 개수
video_fps = video.get(cv2.CAP_PROP_FPS)                 # 동영상 fps

# print(video_count, video_width, video_hight, video_fps)

# 프레임별 저장 폴더
if not os.path.exists('./img/'):
    os.mkdir('./img/')

num = 0
speed = 1

while True:
    ret, img = video.read()

    if ret:
        cv2.imshow('video', img)
        wait_time = int(1000 / (video_fps * speed))
        key = cv2.waitKeyEx(wait_time)

        if video.get(cv2.CAP_PROP_POS_FRAMES) == 1:
            Thumbnail = './img/' + file + '_thumbnail.jpg'
            cv2.imwrite(Thumbnail, img)


        # ESC 영상 종료
        if key == 27:
            break
        
        # Spacebar 이미지 파일로 저장
        elif key == 32:
            filename = './img/' + file + '_' + str(num) + '.jpg'
            
            # 파일 존재시 덮어쓰기 안되게 했습니다.
            while True:
                if os.path.exists(filename) == True:
                    num += 1
                    filename = './img/' + file + '_' + str(num) + '.jpg'
                elif os.path.exists(filename) == False:
                    break
            cv2.imwrite(filename, img)
            num += 1
        

        # R키 누르면 처음 부터
        elif key == 82 or key == 114:
            video.set(cv2.CAP_PROP_POS_FRAMES, 0)

        # S키 누르면 일시 정지 S키 누르면 다시 재생
        elif key == 115 or key == 83:
            while True:
                key = cv2.waitKeyEx(1)

                if key == 115 or key == 83:
                    break
            
        # 좌우방향키 누르면 영상 10초 건너띄기
        elif key == 0x270000:
            video.set(cv2.CAP_PROP_POS_FRAMES, video.get(cv2.CAP_PROP_POS_FRAMES) + 60)

        elif key == 0x250000:
            video.set(cv2.CAP_PROP_POS_FRAMES, video.get(cv2.CAP_PROP_POS_FRAMES) - 60)

        # 상하방향키 누르면 영상 속도조절
        elif key == 0x260000:
            speed += 0.1
        elif key == 0x280000:
            speed -= 0.1
            if speed < 0.1:
                speed = 0.1
            
    else:
        break

cv2.destroyAllWindows()
```
