```python
import cv2, os, sys

txt_path = './1_7_18-00000.txt'
img_path = './1_7_18-00000.jpg'


mouse_point = []
def on_mouse(event , x, y, flags, param):
    
    if event == cv2.EVENT_LBUTTONDOWN:

        print('마우스 왼쪽 눌렀고 좌표는 : ', x, y)
        mouse_point.append([x, y])
    # elif event == cv2.EVENT_RBUTTONDOWN:
    #     print('마우스 오른쪽 눌렀고 좌표는 : ', x, y)

# 원본 txt 파일읽기 - 비율
with open(txt_path, 'r', encoding='utf-8') as f:
    txt_org = f.readlines()

# 원본 읽고 불필요 문자 '\n' 제거
txt_re = []
for _ in txt_org:
    txt_re.append(_.replace('\n', ''))

# 이미지 읽기 - 여기서 좌표 변환 후 박스 그리기
img = cv2.imread(img_path)
img2 = cv2.imread(img_path)
if img is None:
    raise print('이미지 없어 경로 다시확인')

img_w, img_h, img_c = img.shape


# txt 파일 순서를 시각화 하는 변수
txt_num = 0

# opencv식 4개 좌표 
opencv_point = []

# 박스 그리는 위치
for _ in txt_re:
    class_id, x, y, w, h = _.split(' ')
    class_id, x, y, w, h = int(class_id), float(x), float(y), float(w), float(h)

    # yolo식 좌표
    re_x, re_y, re_w, re_h = int(x * img_h), int(y * img_w), int(w * img_h), int(h * img_w)

    # opencv 식 좌표
    x1 = (int(re_x - (re_w/2)), int(re_y + (re_h/2)))
    x2 = (int(re_x + (re_w/2)), int(re_y + (re_h/2)))
    x3 = (int(re_x - (re_w/2)), int(re_y - (re_h/2)))
    x4 = (int(re_x + (re_w/2)), int(re_y - (re_h/2)))

    # opencv식 x1~x4 포인트 좌표
    opencv_point.append([x1, x4])

    thickness = 2
    # 클래스 보고 색상변경
    if class_id == 0:
        # 빨 : 사람
        color = (0, 0, 255)
        a = x3[0]
        b = x3[1]
        cv2.rectangle(img=img, pt1=x1, pt2=x4, color=color, thickness=thickness)

        # 이렇게 하는 이유는 사용자에게 수정할게 있으면 리스트 인덱스를 이용해서 선택하고 클래스명 수정하거나 해당 인덱스번호로 txt파일 삭제
        if b < 0:
            b = x3[1] + 30
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)
        else:
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)
    elif class_id == 1:
        # 파 : 자전거
        color = (255, 0, 0)
        a = x3[0]
        b = x3[1]-2
        cv2.rectangle(img=img, pt1=x1, pt2=x4, color=color, thickness=thickness)
        if b < 0:
            b = x3[1] + 30
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)
        else:
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)

    elif class_id == 2:
        # 녹 : 차
        color = (0, 255, 0)
        a = x3[0]
        b = x3[1]-4
        cv2.rectangle(img=img, pt1=x1, pt2=x4, color=color, thickness=thickness)
        if b < 0:
            b = x3[1] + 30
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)
        else:
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)

    elif class_id == 3:
        # 핑크 : 오토바이
        color = (255, 0, 255)
        a = x3[0]
        b = x3[1]-6
        cv2.rectangle(img=img, pt1=x1, pt2=x4, color=color, thickness=thickness)
        if b < 0:
            b = x3[1] + 30
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)
        else:
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)

    elif class_id == 4:
        # 주황 : 버스
        color = (255, 102, 0)
        a = x3[0]
        b = x3[1]-8

        cv2.rectangle(img=img, pt1=x1, pt2=x4, color=color, thickness=thickness)
        if b < 0:
            b = x3[1] + 30
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)
        else:
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)

    elif class_id == 5:
        # 노랑 : 트럭
        color = (255, 0, 255)
        a = x3[0]
        b = x3[1]-10
        cv2.rectangle(img=img, pt1=x1, pt2=x4, color=color, thickness=thickness)
        if b < 0:
            b = x3[1] + 30
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)
        else:
            cv2.putText(img=img, text=f'{txt_num}', org=(a, b), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=color, thickness=thickness)
    
    txt_num += 1  

# 클래스 개수 적어두기
cv2.putText(img=img, text=f'total : {len(txt_re)}', org=(0,50), fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=1.0, color=(0,0,255), thickness=2)

# 이미지 띄우기
cv2.imshow('label', img)
cv2.imshow('org', img2)

# 마우스 이벤트
cv2.setMouseCallback('label', on_mouse, img)

key = cv2.waitKey()
if key == 27:
    cv2.destroyAllWindows()


# 마우스 포인트 값이 없으면 해당 이미지는 정상이므로 종료
if mouse_point is None:
    sys.exit()
```
