```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import shutil
```

```python
total_txt = []

# txt파일 경로
txt_ls = os.listdir('D:/datas/labels/')

for txt_name in txt_ls:
    with open(f'D:/datas/labels/{txt_name}', 'r', encoding='utf-8') as f:
        txt_org = f.readlines()
        for line in txt_org:
            total_txt.append(f'{txt_name[:-4]} {line}')


yolo_txt = []
for line in total_txt:
    yolo_txt.append(line.split('\n')[0])
data = [row.split() for row in yolo_txt]
```

```python
# 데이터 프레임 적제
df = pd.DataFrame(data, columns=['file_name', 'class', 'w', 'h', 'center_x', 'center_y'])
df = df.set_index('file_name')
df
```

```python
df.describe()
```

```python
dfcounts = pd.DataFrame(df['class'].value_counts())
dfcounts.columns = ['누적 개수']
dfcounts
```

```python
# 한글 폰트
plt.rc('font', family = 'Malgun Gothic') # window
plt.rc('font', size = 12)
plt.rc('axes', unicode_minus = False)
```
